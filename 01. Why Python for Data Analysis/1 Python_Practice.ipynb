```python
#변수 선언
integer_value = 123
float_value = 12.456

#덧셈
sum = integer_value+float_value
print(sum)

#곱셈
multiple = integer_value*float_value
print(multiple)

#나눗셈
division = integer_value/float_value
print(division)

#제곱
integer_value2 = int(float_value)
exponential = integer_value2**2
print(exponential)

#나머지 연산
mod = integer_value%int(float_value)
print(mod)

k=7
p=2
print(k%p)
```

```python
# 1.라이브러리 불러오기

from datetime import datetime as dt
currentDate = dt.now()

# 2.시간, 분 정보 불러오기

hourVar= currentDate.hour
minutesVar = currentDate.minute

print(hourVar)
print(minutesVar)

# 3. 변수의 합 출력

answer=hourVar+minutesVar
print(answer)
```
