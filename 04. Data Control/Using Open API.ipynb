```python
import pandas as pd
import requests
from pandas.io.json import json_normalize
```

```python
# 데이터 접속 주소 입력
url = 'http://kosis.kr/openapi/Param/statisticsParameterData.do?method=getList&apiKey=Mzg4M2JhYWQ2NTRhYzNmNTFlZDI1MzdjYzlkODcwMjg=&itmId=T10+&objL1=ALL&objL2=ALL&objL3=&objL4=&objL5=&objL6=&objL7=&objL8=&format=json&jsonVD=Y&prdSe=M&newEstPrdCnt=1&loadGubun=2&orgId=101&tblId=DT_1G1B002'
# 정상 여부 확인 (200이 정상)
response = requests.get(url)
response
```

```python
# JSON 데이터 획득
json = response.json()
# PandasDataframe 변환
df = json_normalize(json) # json 데이터를 데이터프레임 형식으로 바꾸어줌
df.head()
```

```python
df.to_csv("./data1.csv", encoding='ms949')
```
