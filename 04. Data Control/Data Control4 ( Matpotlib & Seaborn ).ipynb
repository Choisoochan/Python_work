# 데이터 시각화하기

```python
import matplotlib.pyplot as plt
```

### 리스트 생성

```python
xList=list(range(1,20,1))
yList=[]
for i in xList :
   yList.append(i**3)
```

```python
xList
```

```python
yList
```

```python
%matplotlib inline
```

### 1. 차트 그리기

```python
plt.plot(xList,yList)
```

### 2. 차트 옵션 추가 ( 색상 및 선 종류 변경)

```python
plt.plot(xList,yList, 'ko:', lw=2)
```

### 3. 차트의 x, y축 라벨 및 차트 타이틀 생성

```python
plt.plot(xList,yList, 'co-', lw=2)
plt.xlabel('index')
plt.ylabel('value')
plt.title('Test plot')
```

### 4. 차트의 특정 구간 확대

```python
plt.plot(xList,yList, 'mo--', lw=2)
plt.xlim(1,4)
plt.ylim(10,40)
```

### 5. 여러개의 선을 그리고 라벨 붙이기

```python
yList2 = [i**2 for i in xList]
yList3 = [i**2.2 for i in xList]
```

```python
plt.plot(xList,yList, 'mo-', label='zero')
plt.plot(xList,yList2, 'go--', label = 'first')
plt.plot(xList,yList3, 'bo--', label = 'second')
plt.legend(loc=0)
```

### 6. 어노테이션

```python
plt.plot(xList,yList2, 'go--', label = 'first')
plt.plot(xList,yList3, 'bo--', label = 'second')
plt.legend(loc=0)
plt.annotate('point', xy=(8,10),xytext=(10,10),arrowprops=dict(color="blue", width=2))
```

### 7. 차트의 색 및 내용 채우기

```python
plt.plot(xList,yList2, 'go--', label = 'first')
plt.plot(xList,yList3, 'bo--', label = 'second')
plt.grid(True)
plt.xlabel('index')
plt.ylabel('value')
plt.title('Test plot')
```

# 데이터 시각화 고급

```python
# 서브플롯 그리기
```

```python
xList=list(range(1,11,1))
yList2 = [i**2 for i in xList]
yList3 = [i**2.2 for i in xList]

fig = plt.figure(figsize = (10,5))
ax1 = fig.add_subplot(211)
plt.plot(xList,yList2, 'bo-' ,label = '1st')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('subplot')
ax2 = plt.subplot(212)
plt.plot(xList,yList3, 'go-', label='2nd')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

```python
# 이중축 활용하기
```

```python
fig = plt.figure(figsize = (10,5))
ax1 = fig.add_subplot(111)
plt.plot(xList,yList2, 'bo-' ,label = '1st')
plt.grid(True)
plt.legend(loc=2)
plt.ylabel('value')
plt.title('subplot')
ax2 = ax1.twinx()
plt.plot(xList,yList3, 'go-', label='2nd')
plt.legend(loc=4)
plt.xlabel('index')
plt.ylabel('value')
```

실습문제 02. featuresData를 활용하여 2015년도 자료만  
활용하여 x축:week 이중축으로 qty, pro_percent를 차트로 그리시오

```python
featuresData = pd.read_csv('../dataset/feature_regression_example.csv')
```

```python
featuresData2015 = featuresData[(featuresData["YEARWEEK"] >= 201501) & (featuresData["YEARWEEK"] <= 201552)]
```

```python
featuresData2015Week = featuresData2015["WEEK"]
```

```python
featuresData2015Qty = featuresData2015["QTY"]
```

```python
featuresData2015Propercent = featuresData2015["PRO_PERCENT"]
```

```python
fig = plt.figure(figsize = (10,5))

ax1 = fig.add_subplot(111)
plt.plot(featuresData2015Week,featuresData2015Qty, 'bo--')
plt.grid(True)
plt.ylabel('qty')
plt.legend(loc=2)

ax2 = ax1.twinx()
plt.plot(featuresData2015Week,featuresData2015Propercent, 'g')
plt.ylabel('promotion')
plt.title('subplot')
plt.legend(loc=1)
```

# Seaborn 활용

```python
# matplotlib를 사용하여 고급화한 시각화 라이브러리
```

```python
import seaborn as sns
```

```python
# 'tips' 는 웹상에서 제공하는 기본 데이터임
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# countplot : 특정 컬럼 값의 개수를 확인하기위함
```

```python
sns.countplot(x="day", data=tips)
```

```python
# distplot : 특정 컬럼 값의 분보도를 확인하기 위한 차트
```

```python
sns.distplot(tips['total_bill'])
```

```python
# jointplot : 2차원 실수형 데이터를 시각화
```

```python
sns.jointplot(x = 'total_bill', y='tip', data = tips, kind='kde')
```

```python
# pairplot : 3차원 이상데이터 시각화
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex')
```

```python
# barplot : 데이터의 평균 및 실제범위를 시각화
```

```python
# 검은 선은 표준편차를 의미함
```

```python
sns.barplot(x='time', y='total_bill', data = tips)
```

```python
tips.groupby(['time'])['total_bill'].describe()
```

```python
# boxplot : 정해진 카테고리 기준에 어떤 범위의 분포를 보여주는지 차트화함
```

```python
sns.boxplot(x='time', y='total_bill', data=tips)
```

```python
# corr : 숫자형 데이터에 대해서만 시각화를 제공
```

```python
tips.corr()
```

```python
# Heatmap : 자료에 대한 상관관계를 heatmap으로 표현, 색상이 연할수록 연관성이 높다
```

```python
tc = tips.corr()
sns.heatmap(tc, annot = True)
```

```python
flight = sns.load_dataset('flights')
fp = flight.pivot_table(index='month',columns='year',values='passengers')
fp.head()
```

```python
# 연도가 증가함에 따라 승객수 증가, 여름에 집중
```

```python
sns.heatmap(fp)
```

```python
# lmplot : 두 데이터의 상관관계를 linear한 형태로 뽑아내는 차트
```

```python
sns.lmplot(x='total_bill', y='tip', data = tips)
sns.lmplot(x='total_bill', y='tip', data = tips, hue = 'sex', markers = ['o','v'])
```
