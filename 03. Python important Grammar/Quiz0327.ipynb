```python
from isoweek import Week

def preWeek(inputYearWeek,gapWeek):  

    # step 01-------------------------------------------------------------------------------------
    # InputYearWeek의 Index를 사용해야 하기 때문에 어떤 타입이 들어와도 str타입으로 형변환 
    # gapWeek를 어떤 타입이 들어와도 연산값으로 사용해야 하므로 int타입으로 형변환
    # --------------------------------------------------------------------------------------------
    
    currentYearWeek = str(inputYearWeek)
    gapWeek = int(gapWeek)
    
    
    # step 02-------------------------------------------------------------------------------------
    # inputYearWeek를 Index로 변환하여 저장한 currentYearWeek를 연산에 사용하기 위해 Year와 Week로
    # 분리해준 뒤 연산값으로 쓰기 위해 int타입으로 형변환
    # --------------------------------------------------------------------------------------------
    
    currentYear = int(currentYearWeek[0:4])
    currentWeek = int(currentYearWeek[4:])
    
    
    # step 03-------------------------------------------------------------------------------------
    # 연산 후 반환되는 값을(최종 출력해야하는 값) returnWeek 변수에 저장 
    # --------------------------------------------------------------------------------------------
    
    returnWeek = currentWeek - gapWeek
    
    
    # step 04-------------------------------------------------------------------------------------
    # retrunWeek가 음수가 되는 경우 반환되는 값의 출력이 정상적인 포맷을(주차정보는 음수가없음) 
    # 가지지 않기 때문에 올바른 포맷을 띄게 하기 위해 조건을 설정
    # --------------------------------------------------------------------------------------------
    
    if(returnWeek > 0):
        returnWeek = returnWeek
    else:     
        while(returnWeek<=0):
            currentYear = currentYear-1

            maxWeek = Week.last_week_of_year(currentYear).week        
            returnWeek = returnWeek+maxWeek

        
    # step 05-------------------------------------------------------------------------------------
    # week 정보가 0~10미만 일경우 ex)8 -> 08 로 출력하기 위한 조건 
    # --------------------------------------------------------------------------------------------
    
    returnWeekLen = len(str(returnWeek))    
        
    if(returnWeekLen<2):
        returnWeek = '0'+returnWeek
    else:
        returnWeek = returnWeek
    
    # step 06-------------------------------------------------------------------------------------
    # Year정보와 Week정보는 현재 Int타입으로 출력시에는 str타입으로 출력해야하기 때문에
    # str타입으로 Year정보와 Week정보를 저장 후 더하여 함수의 리턴값으로 지정
    # --------------------------------------------------------------------------------------------
    
    returnYearWeek = str(currentYear)+str(returnWeek)
    
    return returnYearWeek


preWeek(201930, 500)
```
